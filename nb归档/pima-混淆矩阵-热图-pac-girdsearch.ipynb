# feature extraction


```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
pima.head()
```

```python
#df[]tuple
X_features=pima[["Glucose","Insulin","BMI","Age"]]
```

# starndardization

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaledX=StandardScaler().fit_transform(X_features)
```

```python
scaledX #ndarray 
```

# ndarray dataframe

```python
dfX=pd.DataFrame(data=scaledX,columns=X_features.columns)
dfX.head()
```

```python
dfy=pima.Outcome#xydf
```

```python
# train my model
#its a binary classification
```

```python
from sklearn.model_selection import train_test_split

train_X,test_X,train_y,test_y=train_test_split(dfX,dfy,random_state=10,test_size=0.2)
```

# 

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
#https://www.jianshu.com/p/284581d9b189
#kfoldtrain test split

from sklearn.linear_model import LogisticRegression#
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC


```

```python
models = []
models.append(("LR", LogisticRegression())) #
models.append(("NB", GaussianNB())) # 
models.append(("KNN", KNeighborsClassifier())) #K
models.append(("DT", DecisionTreeClassifier())) #
models.append(("SVM", SVC())) # 
print(models)
#modelslistlistlen
```

# cross validation

```python
results = []
names = []
for name, themodel in models:
    kfold = KFold(n_splits=10, random_state=22)
    cv_result = cross_val_score(themodel, train_X,train_y, cv = kfold,scoring="accuracy")
    names.append(name)
    results.append(cv_result)
#the model  estimator

for i in range(len(names)):
    print(names[i], results[i].mean)
```

# 

```python
ax = sns.boxplot(data = results)
ax.set_xticklabels(names)
```

# KPCAXsvc

```python
from sklearn.decomposition import KernelPCA
```

```python
kpca=KernelPCA(n_components =2,kernel ="rbf")
#2
```

# fit transform transform

```python
X_train_pca = kpca.fit_transform(train_X)
X_test_pca = kpca.transform(test_X)
```

# 

```python
plt.figure(figsize=(10,8))
plt.scatter(X_train_pca[:,0], X_train_pca[:,1],c=train_y,cmap='plasma')
plt.xlabel("First principal component")
plt.ylabel("Second principal component")
```

# SVC

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix

classifier = SVC(kernel = 'rbf')
classifier.fit(X_train_pca, train_y)
```

```python
y_pred = classifier.predict(X_test_pca)
cm = confusion_matrix(test_y, y_pred)
print(cm)
sns.heatmap(data=cm,annot=True)

print(classification_report(test_y, y_pred))
```

# 

```python
classifier.get_params()
```

```python
#
from sklearn.model_selection import GridSearchCV
param_grid = {'C':[0.1, 1, 10, 100], 'gamma':[1, 0.1, 0.01, 0.001]}
grid = GridSearchCV(SVC(),param_grid,refit=True,verbose = 10)
grid.fit(X_train_pca, train_y)
```

#
