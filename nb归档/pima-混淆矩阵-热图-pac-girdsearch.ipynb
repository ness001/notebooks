```python
#注意df[]结构中传导一个tuple，所以有两对【】
X_features=pima[["Glucose","Insulin","BMI","Age"]]
```

# starndardization

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaledX=StandardScaler().fit_transform(X_features)
```

```python
scaledX #返回的是ndarray 不能直接使用
```

# 把ndarray转化为 dataframe

```python
dfX=pd.DataFrame(data=scaledX,columns=X_features.columns)
dfX.head()
```

```python
dfy=pima.Outcome#xy都是df结构
```

```python
# train my model
#its a binary classification
```

```python
from sklearn.model_selection import train_test_split

train_X,test_X,train_y,test_y=train_test_split(dfX,dfy,random_state=10,test_size=0.2)
```

# 二分类问题例子

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
#https://www.jianshu.com/p/284581d9b189
#kfold通常和交叉验证一起使用，必须之前使用train test split分出验证集

from sklearn.linear_model import LogisticRegression#罗辑回归实际上是分类
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC


```

```python
models = []
models.append(("LR", LogisticRegression())) #逻辑回归
models.append(("NB", GaussianNB())) # 高斯朴素贝叶斯
models.append(("KNN", KNeighborsClassifier())) #K近邻分类
models.append(("DT", DecisionTreeClassifier())) #决策树分类
models.append(("SVM", SVC())) # 支持向量机分类
print(models)
#models是list，加的元素是元组，list可以用len来知道个数
```

# cross validation对多个模型评估

```python
results = []
names = []
for name, themodel in models:
    kfold = KFold(n_splits=10, random_state=22)
    cv_result = cross_val_score(themodel, train_X,train_y, cv = kfold,scoring="accuracy")
    names.append(name)
    results.append(cv_result)
#the model 是 estimator

for i in range(len(names)):
    print(names[i], results[i].mean)
```

# 可视化多个模型评估结果

```python
ax = sns.boxplot(data = results)
ax.set_xticklabels(names)
```

# KPCA改造特征值X为svc作准备

```python
from sklearn.decomposition import KernelPCA
```

```python
kpca=KernelPCA(n_components =2,kernel ="rbf")
#似乎是将它投射到2维？
```

# fit transform和 transform

```python
X_train_pca = kpca.fit_transform(train_X)
X_test_pca = kpca.transform(test_X)
```

# ？

```python
plt.figure(figsize=(10,8))
plt.scatter(X_train_pca[:,0], X_train_pca[:,1],c=train_y,cmap='plasma')
plt.xlabel("First principal component")
plt.ylabel("Second principal component")
```

# SVC模型评估，分类报告，混淆矩阵

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix

classifier = SVC(kernel = 'rbf')
classifier.fit(X_train_pca, train_y)
```

```python
y_pred = classifier.predict(X_test_pca)
cm = confusion_matrix(test_y, y_pred)
print(cm)
sns.heatmap(data=cm,annot=True)

print(classification_report(test_y, y_pred))
```

# 获得模型的参数

```python
classifier.get_params()
```

# 网格搜索寻找最优参数

```python
#利用网格搜索找到最优
from sklearn.model_selection import GridSearchCV
param_grid = {'C':[0.1, 1, 10, 100], 'gamma':[1, 0.1, 0.01, 0.001]}
grid = GridSearchCV(SVC(),param_grid,refit=True,verbose = 10)
grid.fit(X_train_pca, train_y)
```

# 直接用最优参数进行模型评估

```python
grid_predictions =grid.predict(X_test_pca)
```

```python
print(classification_report(test_y,grid_predictions))
```
