## make subplots(mixed)

```python
x1=np.arange(0.0,5.0,0.1)
y1=np.sin(2*np.pi*x1)
```

```python
x2=np.arange(0.0,5.0,0.02)
y2=np.cos(2*np.pi*x2)
```

```python
fig,axes_list=plt.subplots(2,1)
```

```python
axes_list[0].plot(x1,y1,'bo')
axes_list[1].plot(x2,y2,'r--')
```

```python
fig
```

## 面向过程仅使用plt的编程方法绘制子图

```python
plt.subplot(2,1,1)
plt.plot(x1,y1)
plt.title('subplot1')
plt.legend(['subplot1'])
plt.xlabel('x label')
plt.ylabel('y label')
           
           
plt.subplot(2,1,2)
plt.scatter(x2,y2,color='r',alpha=0.5,s=1)
plt.title('subplot1')
plt.legend(['subplot1'])
plt.xlabel('x label')
plt.ylabel('y label')
```
