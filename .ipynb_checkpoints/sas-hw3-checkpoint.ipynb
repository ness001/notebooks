```python
#煤球单位统一化
df.loc[df['eu']=='斤','et']=df.loc[df['eu']=='斤','et']*500/1000
#经过搜索得知，一个煤球约为1.5kg左右，所以对煤球个数单位做标准化
df.loc[df['eu']=='个','et']=df.loc[df['eu']=='个','et']*1.5 
#1吨=1000kg
df.loc[df['euo']=='吨','et']=df.loc[df['euo']=='吨','et']*1000 
```

```python
##对煤块单位统一到公斤,一个煤块约为2kg
df.loc[df['cbu']=='斤','cbt']=df.loc[df['cbu']=='斤','cbt']*500/1000
df.loc[df['cbu']=='个','cbt']=df.loc[df['cbu']=='个','cbt']*2 
df.loc[df['cbuo']=='吨','cbt']=df.loc[df['cbuo']=='吨','cbt']*1000 
```

```python
#将汽油统一到升,一公斤汽油1.4升左右,公升等于升,
df.loc[df['gu']=='公斤','gt']=df.loc[df['gu']=='公斤','gt']*1.4
df.loc[df['gu']=='斤','gt']=df.loc[df['gu']=='斤','gt']*500/1000*1.4 
#一瓶矿泉水大约是0.35升
df.loc[df['guo']=='矿泉水瓶装','gt']=df.loc[df['guo']=='矿泉水瓶装','gt']*0.35 
```

```python
#将柴油单位统一到升,1kg柴油约等于1.12升
df.loc[df['du']=='公斤','dt']=df.loc[df['du']=='公斤','dt']*1.12
df.loc[df['du']=='斤','dt']=df.loc[df['du']=='斤','dt']*500/1000*1.12
```

```python
#将液化气单位,公升和斤,统一到公斤(千克),一升液化气约等于一公斤,所以不做改变
df.loc[df['lgu']=='斤','lgt']=df.loc[df['lgu']=='斤','lgt']*500/1000
```

```python
#将粪便统一到公斤
df.loc[df['fu']=='斤','ft']=df.loc[df['fu']=='斤','ft']*500/1000 
```

```python
#将秸秆单位统一到公斤
df.loc[df['jgu']=='斤','jgt']=df.loc[df['jgu']=='斤','jgt']*500/1000 
```

```python
#将木柴单位统一到公斤
df.loc[df['fwu']=='斤','fwt']=df.loc[df['fwu']=='斤','fwt']*500/1000 
```

```python
#将其他能源单位的斤统一到公斤,立方米统一到升
df.loc[df['opu']=='斤','opt']=df.loc[df['opu']=='斤','opt']*500/1000 
df.loc[df['opu']=='立方米','opt']=df.loc[df['opu']=='立方米','opt']*1000 
```

```python
df.to_csv('power_cleaned.csv',encoding='utf8')
```
